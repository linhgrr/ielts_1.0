```python
import tensorflow as tf
import pandas as pd
import numpy as np
from transformer import Transformer
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from Layers.mask import create_masks
import pickle
import os
import re
import string
import matplotlib.pyplot as plt
from keras.saving import register_keras_serializable
from tensorflow.keras.metrics import SparseTopKCategoricalAccuracy
```

```python
def normalize_text(text):
    text = text.lower()  # Chuyển thành chữ thường
    text = text.translate(str.maketrans('', '', string.punctuation))  # Xóa dấu câu
    text = re.sub(r'[^a-zA-Z0-9\sÀÁÂÃÈÉÊÌÍÒÓÔÕÙÚĂĐĨŨƠàáâãèéêìíòóôõùúăđĩũơƯĂẠẢẤẦẨẪẬẮẰẲẴẶẸẺẼỀỀỂưăạảấầẩẫậắằẳẵặẹẻẽềềểỄỆỈỊỌỎỐỒỔỖỘỚỜỞỠỢỤỦỨỪễệỉịọỏốồổỗộớờởỡợụủứừỬỮỰỲỴÝỶỸửữựỳỵỷỹ]', '', text)  # Giữ chữ cái và số
    text = ' '.join(text.split())  # Xóa khoảng trắng thừa
    return text
```

```python
def filter_text_length(data):
    min_length = 2  # Tối thiểu 2 từ
    max_length = 50  # Tối đa 50 từ
    data = data[data['en'].apply(lambda x: min_length <= len(x.split()) <= max_length)]
    data = data[data['vi'].apply(lambda x: min_length <= len(x.split()) <= max_length)]

    return data
```

```python
def show_data(data):
    data['en_length'] = data['en'].apply(lambda x: len(x.split()))
    data['vi_length'] = data['vi'].apply(lambda x: len(x.split()))
    plt.hist(data['en_length'], bins=50, alpha=0.5, label='Tiếng Anh')
    plt.hist(data['vi_length'], bins=50, alpha=0.5, label='Tiếng Việt')
    plt.legend()
    plt.show()
```

```python
max_vocab_size=50000
max_length=50
eng_tokenizer = Tokenizer(num_words=max_vocab_size, oov_token="<OOV>")
vie_tokenizer = Tokenizer(num_words=max_vocab_size, oov_token="<OOV>")
```

```python
def load_data():
    train_data = pd.read_csv('Data/train.csv')
    valid_data = pd.read_csv('Data/valid.csv')

    # Loại bỏ giá trị NaN
    train_data.dropna(inplace=True)
    valid_data.dropna(inplace=True)

    train_data = filter_text_length(train_data)
    valid_data = filter_text_length(valid_data)
    # Áp dụng chuẩn hóa văn bản
    train_data['en'] = train_data['en'].apply(normalize_text)
    train_data['vi'] = train_data['vi'].apply(normalize_text)
    valid_data['en'] = valid_data['en'].apply(normalize_text)
    valid_data['vi'] = valid_data['vi'].apply(normalize_text)

    # Fit tokenizer trên dữ liệu đã chuẩn hóa
    eng_tokenizer.fit_on_texts(train_data['en'].values)
    vie_tokenizer.fit_on_texts(train_data['vi'].values)

    # Thêm token đặc biệt cho tiếng Việt
    vie_tokenizer.word_index['<start>'] = len(vie_tokenizer.word_index) + 1
    vie_tokenizer.word_index['<end>'] = len(vie_tokenizer.word_index) + 1
    vie_tokenizer.index_word[vie_tokenizer.word_index['<start>']] = '<start>'
    vie_tokenizer.index_word[vie_tokenizer.word_index['<end>']] = '<end>'

    return train_data, valid_data
```

```python
def preprocess_data(data):
      eng_sequences = eng_tokenizer.texts_to_sequences(data['en'].astype(str).values)
      vie_sequences = vie_tokenizer.texts_to_sequences(data['vi'].astype(str).values)

      vie_sequences = [[vie_tokenizer.word_index['<start>']] + seq + [vie_tokenizer.word_index['<end>']]
                        for seq in vie_sequences]
      eng_sequences = pad_sequences(eng_sequences, maxlen=max_length, padding='post')
      vie_sequences = pad_sequences(vie_sequences, maxlen=max_length, padding='post')

      return eng_sequences, vie_sequences
```

```python
@register_keras_serializable()
class CustomSchedule(tf.keras.optimizers.schedules.LearningRateSchedule):
    def __init__(self, d_model, warmup_steps=4000):
        super(CustomSchedule, self).__init__()
        self.d_model = float(d_model)
        self.warmup_steps = warmup_steps

    def __call__(self, step):
        step = tf.cast(step, tf.float32)
        arg1 = tf.math.rsqrt(step)
        arg2 = step * (self.warmup_steps ** -1.5)
        return tf.math.rsqrt(self.d_model) * tf.math.minimum(arg1, arg2)

    def get_config(self):
        return {
            'd_model': self.d_model,
            'warmup_steps': self.warmup_steps
        }

    @classmethod
    def from_config(cls, config):
        return cls(**config)

learning_rate = CustomSchedule(embedding_dim)
optimizer = tf.keras.optimizers.Adam(learning_rate, beta_1=0.9, beta_2=0.98, epsilon=1e-9)
```

```python
import tensorflow as tf
from keras.saving import register_keras_serializable

@register_keras_serializable()
class CustomSchedule(tf.keras.optimizers.schedules.LearningRateSchedule):
    def __init__(self, d_model, warmup_steps=4000):
        super(CustomSchedule, self).__init__()
        self.d_model = float(d_model)  # Ensure d_model is a float to match the config
        self.warmup_steps = warmup_steps

    def __call__(self, step):
        # Common learning rate schedule for transformers
        step = tf.cast(step, tf.float32)
        arg1 = tf.math.rsqrt(step)
        arg2 = step * (self.warmup_steps ** -1.5)
        return tf.math.rsqrt(self.d_model) * tf.math.minimum(arg1, arg2)

    def get_config(self):
        # Must return the same config keys as in the error message
        return {
            'd_model': self.d_model,
            'warmup_steps': self.warmup_steps
        }
```

```python
def sample_with_temperature(predictions, temperature=0.7):
    predictions = np.asarray(predictions).astype('float64')
    predictions = np.log(predictions + 1e-10) / temperature  # Apply temperature
    exp_preds = np.exp(predictions)
    predictions = exp_preds / np.sum(exp_preds)  # Normalize
    return np.random.choice(len(predictions), p=predictions)
```

```python
def translate(text):
    # Chuẩn hóa văn bản đầu vào
    text = normalize_text(text)
    eng_sequence = eng_tokenizer.texts_to_sequences([text])
    eng_sequence = pad_sequences(eng_sequence, maxlen=max_length, padding='post')

    decoded_sentence = ''
    start_token = vie_tokenizer.word_index['<start>']
    end_token = vie_tokenizer.word_index['<end>']

    target_sequence = np.zeros((1, 1), dtype=np.int32)
    target_sequence[0, 0] = start_token

    for _ in range(max_length):
        enc_padding_mask, look_ahead_mask, dec_padding_mask = create_masks(eng_sequence, target_sequence)
        predictions = model.predict([eng_sequence, target_sequence, enc_padding_mask, look_ahead_mask, dec_padding_mask])

        predicted_id = np.argmax(predictions[0, -1, :])

        if predicted_id == end_token:
            break

        print(vie_tokenizer.index_word[predicted_id])

        decoded_sentence += vie_tokenizer.index_word[predicted_id] + ' '
        target_sequence = np.append(target_sequence, [[predicted_id]], axis=1)

    return decoded_sentence.strip()
```

```python
import numpy as np

def translate_beam_search(text, beam_width=5):
    # Chuẩn hóa văn bản đầu vào
    text = normalize_text(text)
    eng_sequence = eng_tokenizer.texts_to_sequences([text])
    eng_sequence = pad_sequences(eng_sequence, maxlen=max_length, padding='post')

    start_token = vie_tokenizer.word_index['<start>']
    end_token = vie_tokenizer.word_index['<end>']

    # Khởi tạo với chuỗi bắt đầu chỉ chứa <start>
    sequences = [(np.array([[start_token]]), 0.0)]

    for _ in range(max_length):
        all_candidates = []
        for seq, score in sequences:
            # Nếu chuỗi đã kết thúc, không mở rộng thêm
            if seq[0, -1] == end_token:
                all_candidates.append((seq, score))
                continue

            enc_padding_mask, look_ahead_mask, dec_padding_mask = create_masks(eng_sequence, seq)
            predictions = model.predict([eng_sequence, seq, enc_padding_mask, look_ahead_mask, dec_padding_mask])
            # Lấy dự đoán cho token cuối cùng (dạng xác suất)
            preds = predictions[0, -1, :]  # shape: (vocab_size,)

            # Lấy chỉ số của top beam_width từ các xác suất cao nhất
            top_indices = np.argsort(preds)[-beam_width:][::-1]

            for word_id in top_indices:
                prob = preds[word_id]
                # Bỏ qua nếu xác suất bằng 0 để tránh lỗi log(0)
                if prob == 0:
                    continue
                new_score = score + np.log(prob)
                # Nếu là token kết thúc, không mở rộng thêm
                if word_id == end_token:
                    candidate_seq = seq
                else:
                    candidate_seq = np.concatenate([seq, np.array([[word_id]])], axis=1)
                all_candidates.append((candidate_seq, new_score))

        # Sắp xếp các candidate theo điểm số và chọn beam_width chuỗi có điểm cao nhất
        sequences = sorted(all_candidates, key=lambda x: x[1], reverse=True)[:beam_width]

        # Nếu tất cả các chuỗi đều kết thúc, dừng vòng lặp
        if all(seq[0, -1] == end_token for seq, _ in sequences):
            break

    # Lấy chuỗi có điểm cao nhất
    best_seq = sequences[0][0][0]  # Chuyển sang dạng 1D
    decoded_sentence = ' '.join([vie_tokenizer.index_word[word_id] for word_id in best_seq if word_id not in [start_token, end_token]])

    return decoded_sentence.strip()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import string
from collections import Counter
from tensorflow.keras.preprocessing.text import Tokenizer
import random

# ------------ 1. KIỂM TRA THỐNG KÊ CƠ BẢN ------------
def basic_data_stats(data):
    """Hiển thị thống kê cơ bản về dữ liệu"""
    print(f"Tổng số mẫu: {len(data)}")
    print(f"Các cột: {data.columns.tolist()}")

    # Kiểm tra giá trị null
    null_counts = data.isnull().sum()
    print("\nSố lượng giá trị null trên mỗi cột:")
    print(null_counts)

    # Kiểm tra giá trị trùng lặp
    duplicate_count = data.duplicated().sum()
    print(f"\nSố lượng mẫu trùng lặp: {duplicate_count}")

    # Thống kê độ dài
    data['en_length'] = data['en'].apply(lambda x: len(str(x).split()))
    data['vi_length'] = data['vi'].apply(lambda x: len(str(x).split()))

    print("\nThống kê độ dài (số từ):")
    print("Tiếng Anh:")
    print(data['en_length'].describe())
    print("\nTiếng Việt:")
    print(data['vi_length'].describe())

    return data

# ------------ 2. KIỂM TRA ĐỘ DÀI CÂU ------------
def plot_sentence_lengths(data):
    """Vẽ biểu đồ phân bố độ dài câu"""
    plt.figure(figsize=(15, 6))

    plt.subplot(1, 2, 1)
    sns.histplot(data['en_length'], bins=50, kde=True)
    plt.title('Phân bố độ dài câu tiếng Anh')
    plt.xlabel('Số từ')
    plt.ylabel('Số lượng câu')

    plt.subplot(1, 2, 2)
    sns.histplot(data['vi_length'], bins=50, kde=True)
    plt.title('Phân bố độ dài câu tiếng Việt')
    plt.xlabel('Số từ')
    plt.ylabel('Số lượng câu')

    plt.tight_layout()
    plt.show()

    # Vẽ biểu đồ scatter để so sánh độ dài giữa tiếng Anh và tiếng Việt
    plt.figure(figsize=(10, 6))
    plt.scatter(data['en_length'], data['vi_length'], alpha=0.5)
    plt.title('So sánh độ dài câu Anh-Việt')
    plt.xlabel('Độ dài câu tiếng Anh (số từ)')
    plt.ylabel('Độ dài câu tiếng Việt (số từ)')

    # Thêm đường hồi quy tuyến tính
    z = np.polyfit(data['en_length'], data['vi_length'], 1)
    p = np.poly1d(z)
    plt.plot(data['en_length'], p(data['en_length']), "r--", alpha=0.8)

    plt.tight_layout()
    plt.show()

# ------------ 3. KIỂM TRA PHÂN BỐ TỪ VỰNG ------------
def check_vocabulary_distribution(data, max_vocab=50):
    """Kiểm tra phân bố từ vựng thường gặp nhất"""
    # Tạo counters từ vựng
    en_words = ' '.join(data['en'].astype(str)).lower().split()
    vi_words = ' '.join(data['vi'].astype(str)).lower().split()

    en_counter = Counter(en_words)
    vi_counter = Counter(vi_words)

    # Hiển thị số lượng từ duy nhất
    print(f"Số lượng từ duy nhất tiếng Anh: {len(en_counter)}")
    print(f"Số lượng từ duy nhất tiếng Việt: {len(vi_counter)}")

    # Vẽ biểu đồ top từ thường gặp
    plt.figure(figsize=(15, 10))

    plt.subplot(2, 1, 1)
    top_en = dict(en_counter.most_common(max_vocab))
    plt.bar(top_en.keys(), top_en.values())
    plt.xticks(rotation=90)
    plt.title(f'Top {max_vocab} từ tiếng Anh thường gặp nhất')

    plt.subplot(2, 1, 2)
    top_vi = dict(vi_counter.most_common(max_vocab))
    plt.bar(top_vi.keys(), top_vi.values())
    plt.xticks(rotation=90)
    plt.title(f'Top {max_vocab} từ tiếng Việt thường gặp nhất')

    plt.tight_layout()
    plt.show()

    return en_counter, vi_counter

# ------------ 4. KIỂM TRA TOKENIZER ------------
def check_tokenizer_coverage(data, max_vocab_size=50000):
    """Kiểm tra độ phủ của tokenizer"""
    # Tạo và fit tokenizers
    eng_tokenizer = Tokenizer(num_words=max_vocab_size, oov_token="<OOV>")
    vie_tokenizer = Tokenizer(num_words=max_vocab_size, oov_token="<OOV>")

    eng_tokenizer.fit_on_texts(data['en'].astype(str).values)
    vie_tokenizer.fit_on_texts(data['vi'].astype(str).values)

    # Kiểm tra kích thước từ điển thực tế
    print(f"Kích thước từ điển tiếng Anh: {len(eng_tokenizer.word_index)}")
    print(f"Kích thước từ điển tiếng Việt: {len(vie_tokenizer.word_index)}")

    # Tính tỷ lệ OOV
    eng_sequences = eng_tokenizer.texts_to_sequences(data['en'].astype(str).values)
    vie_sequences = vie_tokenizer.texts_to_sequences(data['vi'].astype(str).values)

    eng_total_tokens = sum(len(seq) for seq in eng_sequences)
    vie_total_tokens = sum(len(seq) for seq in vie_sequences)

    eng_oov_count = sum(1 for seq in eng_sequences for token in seq if token == 1)  # OOV token index = 1
    vie_oov_count = sum(1 for seq in vie_sequences for token in seq if token == 1)

    print(f"Tỷ lệ OOV tiếng Anh: {eng_oov_count/eng_total_tokens:.4f}")
    print(f"Tỷ lệ OOV tiếng Việt: {vie_oov_count/vie_total_tokens:.4f}")

    return eng_tokenizer, vie_tokenizer

# ------------ 5. KIỂM TRA DẤU TIẾNG VIỆT ------------
def check_vietnamese_diacritics(data):
    """Kiểm tra dấu tiếng Việt trong dữ liệu"""
    # Mẫu regex cho các chữ cái tiếng Việt có dấu
    vietnamese_chars = r'[àáảãạăắằẳẵặâấầẩẫậèéẻẽẹêếềểễệìíỉĩịòóỏõọôốồổỗộơớờởỡợùúủũụưứừửữựỳýỷỹỵđ]'

    # Đếm số câu có dấu
    has_diacritics = data['vi'].astype(str).apply(lambda x: bool(re.search(vietnamese_chars, x.lower())))
    diacritic_count = has_diacritics.sum()

    print(f"Số câu tiếng Việt có dấu: {diacritic_count} ({diacritic_count/len(data):.2%})")

    # Hiển thị một số mẫu không có dấu
    if len(data) - diacritic_count > 0:
        print("\nMột số mẫu tiếng Việt không có dấu:")
        no_diacritic_samples = data[~has_diacritics]['vi'].head(5).tolist()
        for i, sample in enumerate(no_diacritic_samples):
            print(f"{i+1}. {sample}")

    return has_diacritics

# ------------ 6. KIỂM TRA CHẤT LƯỢNG DỮ LIỆU ------------
def sample_translation_pairs(data, n=10):
    """Lấy mẫu ngẫu nhiên các cặp câu để kiểm tra chất lượng"""
    sample_indices = random.sample(range(len(data)), n)

    print("Mẫu cặp câu dịch (để kiểm tra chất lượng):")
    for i, idx in enumerate(sample_indices):
        print(f"\nCặp {i+1}:")
        print(f"EN: {data.iloc[idx]['en']}")
        print(f"VI: {data.iloc[idx]['vi']}")

    return sample_indices

# ------------ 7. PHÁT HIỆN DỮ LIỆU BẤT THƯỜNG ------------
def detect_anomalies(data):
    """Phát hiện dữ liệu bất thường"""
    # Tỷ lệ độ dài bất thường
    length_ratio = data['vi_length'] / data['en_length']

    plt.figure(figsize=(10, 6))
    sns.histplot(length_ratio, bins=50, kde=True)
    plt.title('Phân bố tỷ lệ độ dài (Việt/Anh)')
    plt.xlabel('Tỷ lệ độ dài')
    plt.axvline(x=length_ratio.mean(), color='r', linestyle='--', label=f'Giá trị trung bình: {length_ratio.mean():.2f}')
    plt.legend()
    plt.show()

    # Xác định cặp câu có tỷ lệ độ dài bất thường
    anomaly_threshold = 3  # Tỷ lệ độ dài lớn hơn 3 lần hoặc nhỏ hơn 1/3
    length_anomalies = data[(length_ratio > anomaly_threshold) | (length_ratio < 1/anomaly_threshold)]

    print(f"Số cặp câu có tỷ lệ độ dài bất thường: {len(length_anomalies)} ({len(length_anomalies)/len(data):.2%})")

    if len(length_anomalies) > 0:
        print("\nMột số mẫu có tỷ lệ độ dài bất thường:")
        sample_anomalies = length_anomalies.head(5)
        for i, (_, row) in enumerate(sample_anomalies.iterrows()):
            print(f"\nMẫu {i+1} (tỷ lệ: {row['vi_length']/row['en_length']:.2f}):")
            print(f"EN ({row['en_length']} từ): {row['en']}")
            print(f"VI ({row['vi_length']} từ): {row['vi']}")

    return length_anomalies

# ------------ 8. RUN TẤT CẢ CÁC KIỂM TRA ------------
def run_all_checks(data_path):
    """Chạy tất cả các kiểm tra trên dữ liệu"""
    # Đọc dữ liệu
    print(f"Đọc dữ liệu từ {data_path}...")
    data = pd.read_csv(data_path)

    # Chạy các kiểm tra
    print("\n=============== THỐNG KÊ CƠ BẢN ===============")
    data = basic_data_stats(data)

    print("\n=============== PHÂN BỐ ĐỘ DÀI CÂU ===============")
    plot_sentence_lengths(data)

    print("\n=============== PHÂN BỐ TỪ VỰNG ===============")
    en_counter, vi_counter = check_vocabulary_distribution(data, max_vocab=20)

    print("\n=============== KIỂM TRA TOKENIZER ===============")
    eng_tokenizer, vie_tokenizer = check_tokenizer_coverage(data)

    print("\n=============== KIỂM TRA DẤU TIẾNG VIỆT ===============")
    has_diacritics = check_vietnamese_diacritics(data)

    print("\n=============== MẪU CẶP CÂU DỊCH ===============")
    sample_indices = sample_translation_pairs(data, n=5)

    print("\n=============== PHÁT HIỆN DỮ LIỆU BẤT THƯỜNG ===============")
    anomalies = detect_anomalies(data)

    print("\nHoàn tất kiểm tra dữ liệu!")
    return data

# Gọi hàm để chạy kiểm tra
# Thay 'Data/train.csv' bằng đường dẫn đến file dữ liệu của bạn
if __name__ == "__main__":
    train_data = run_all_checks('Data/train.csv')

    # Bạn cũng có thể kiểm tra dữ liệu validation riêng
    print("\n\n==================== KIỂM TRA DỮ LIỆU VALIDATION ====================")
    valid_data = run_all_checks('Data/valid.csv')
```
